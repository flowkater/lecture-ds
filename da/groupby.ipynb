## Group by - Basic

```python
# data from: 
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data)
df
```

```python
df["Team"].unique()
```

```python
df.groupby("Team")["Points"].sum()
```

```python
df.groupby("Team")["Points"].sum().tolist()
```

```python
h_index = df.groupby(["Team", "Year"])["Points"].sum()
h_index
```

```python
h_index.index
```

```python
h_index.unstack().fillna(0)
```

```python
h_index.swaplevel().sort_index(level=0)
```

```python
h_index
```

```python
h_index.sum(level=0)
```

```python
h_index.sum(level=1)
```

### Groupby - grouped

```python
grouped = df.groupby("Team")
```

```python
for data in grouped:
    print(data)
```

```python
for name, group in grouped:
    print(name)
    print(group)
```

```python
for name, group in grouped:
    print(type(name))
    print(type(group))
```

```python
grouped.agg(min)
```

```python
grouped.agg(max)
```

```python
import numpy as np
grouped.agg(np.mean)
```

```python
grouped.agg([np.sum, np.mean, np.std])
```

```python
grouped["Points"].agg([np.sum, np.mean, np.std]).fillna(0)
```

```python
df_phone = pd.read_csv('./data/phone_data.csv', parse_dates=["date"])
print(df_phone.shape)
df_phone.head()
```

```python
df_phone.groupby('month')['duration'].sum()
```

```python
df_phone.groupby('month')['duration'].mean()
```

```python
df_phone[df_phone["item"] == "call"].groupby('month')['duration'].sum()
```

```python
df_phone.groupby(["month", "item"])["duration"].sum()
```

```python
df_phone.groupby(["month", "item"])["date"].count()
```

```python
df_phone.groupby(["month", "item"])["date"].count().unstack()
```

```python
df_phone.groupby('month', as_index=False).agg({'duration': sum})
```

```python
df_phone.groupby(['month', 'item']).agg({'duration': sum, \
                                         'network_type': 'count', \
                                         'date': 'first'})
```

```python
df_phone.groupby(['month', 'item']).agg({'duration': [min, max, sum], \
                                         'network_type': 'count', \
                                         'date': [min, 'first', 'nunique']})
```

```python
grouped = df_phone.groupby('month').agg({'duration': [min, max, np.mean]})
grouped
```

```python
grouped.columns = grouped.columns.droplevel(level=0)
grouped
```

```python
grouped.rename(columns={"min": "grouped_min", "max": "grouped_max", "mean": "grouped_mean"})
```

```python
grouped = df_phone.groupby('month').agg({'duration': [min, max, np.mean]})
grouped.columns = grouped.columns.droplevel(level=0)
grouped
```

```python
grouped.add_prefix("duration_")
```

## Pivot Table

```python
df_phone['network'].unique()
```

```python
df_phone.pivot_table(["duration"],
                     index=[df_phone.month, df_phone.item],
                     columns=df_phone.network, aggfunc="sum", fill_value=0)
```

```python
df_phone.groupby(['month','item','network']).agg({'duration':sum}).unstack().fillna(0)
```

```python
df_movie = pd.read_csv("./data/movie_rating.csv")
print(df_movie.shape)
df_movie.head()
```

```python
df_movie
```

```python
df_movie.pivot_table(["rating"], index=df_movie.critic, columns=df_movie.title, aggfunc="first", fill_value=0)
```

```python
df_movie.groupby(['critic', 'title']).agg({'rating': 'first'}).unstack().fillna(0)
```

```python
pd.crosstab(index=df_movie.critic, \
            columns=df_movie.title, \
            values=df_movie.rating, \
            aggfunc="first").fillna(0)
```
